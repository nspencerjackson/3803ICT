```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
job_market = pd.read_csv('job_market.csv')
job_market.head()
```

```python
plt.figure(figsize = (15,6))
job_market['Location'].hist()
plt.xticks(rotation = 270)
plt.show()
```

```python
job_market.info()
```

```python
plt.figure(figsize = (6,6))
job_market['HighestSalary'].hist()
plt.title('dd')
plt.show()
```

```python
plt.figure(figsize = (6,6))
job_market['LowestSalary'].hist()
plt.title('dd')
plt.show()
```
