```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Load job market

job_market = pd.read_csv('job_market.csv')
job_market.head(3)
```

```python
%matplotlib inline

plt.figure(figsize = (20,20))
sns.color_palette('husl', 8)
sns.countplot(data = job_market, y = 'Classification', order = job_market['Classification'].value_counts().index)
sns.set(font_scale = 2)
plt.show()
```
